## 数据预处理
### 1.去掉乱码
通常用的空格是 \x20
* \xa0 是拉丁字符的不间断空格符，无法用gbk编码，需要被替换掉
str的 replace函数
### 2.变换大小写：统一变小写
str的lower函数
### 3.分词
list或是split或是jieba等
### 4.建立词典
词频统计：collections.Counter <br>
特殊token一般有： pad,bos,eos,unk(补足，句子开头，句子结尾，没有的单词) <br>
两个词典 idx2token和token2idx <br>
idx2token用列表存就可以，idx作为索引可以获取token <br>
token2idx则用字典存，dict[token] = idx
### 5.载入数据
* 对每个语句的开头和结尾添加 bos 和 eos
* 对不同大小的语句，如何整合成一个mini_batch的数据集？（假设指定固定的语句长度len）<br>
对短的语句，使用**pad填充**，将短的语句pad到len长度，<br>
对长的语句，切片前面len个token.<br>

最后的目的是通过token2idx词典，将一句话根据token转换成一个向量。
